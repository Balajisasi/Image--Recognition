### Importing Libraries

```python
import os
import warnings
warnings.simplefilter('ignore')
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from skimage.io import imread,imshow
from skimage.transform import resize
from skimage.color import rgb2gray
from sklearn.model_selection import train_test_split
```

### Opening the Directories of the images

```python
pew=os.listdir("C:/Users/Yogesh/Desktop/Image-recognition/pewdiepiw/")
jenna=os.listdir("C:/Users/Yogesh/Desktop/Image-recognition/jenna/")
jk=os.listdir("C:/Users/Yogesh/Desktop/picture daya/jk/")
```

### Getting the images from the directories

```python
limit=10
pew_images = [None]*limit
jenna_images = [None]*limit
jk_images = [None]*limit
j=0
for i in pew:
    if(j<limit):
        pew_images[j]=imread("C:/Users/Yogesh/Desktop/picture daya/pewdiepiw/"+i)
        j+=1
    else:
        break

j=0
for i in jk:
    if(j<limit):
        jk_images[j]=imread("C:/Users/Yogesh/Desktop/picture daya/jk/"+i)
        j+=1
    else:
        break
j=0
for i in jenna:
    if(j<limit):
        jenna_images[j]=imread("C:/Users/Yogesh/Desktop/picture daya/jenna/"+i)
        j+=1
    else:
        break
        
```

### Converting the images to grayscale

```python
pew_gray=[None]*limit
jenna_gray=[None]*limit
jk_gray=[None]*limit
j=0

for i in pew:
    if(j<limit):
        pew_gray[j]=rgb2gray(pew_images[j])
        j+=1
    else:
        break
j=0
for i in jenna:
    if(j<limit):
        jenna_gray[j]=rgb2gray(jenna_images[j])
        j+=1
    else:
        break
    

j=0
for i in jk:
    if(j<limit):
        jk_gray[j]=rgb2gray(jk_images[j])
        j+=1
    else:
        break
```

```python
imshow(pew_gray[9])
```

```python
pew_gray[2].shape
```

### Resizing all the images to same size

```python
for j in range(10):
    jk_temp=jk_gray[j]
    jk_gray[j]=resize(jk_temp,(512,512))
for j in range(10):
    pew_temp=pew_gray[j]
    pew_gray[j]=resize(pew_temp,(512,512))
for j in range(10):
    jenna_temp=jenna_gray[j]
    jenna_gray[j]=resize(jenna_temp,(512,512))
```

```python
imshow(pew_gray[2])
```

### Storing the number of images and pixels into new variables

```python
len_of_jk=len(jk_gray)
len_of_jenna=len(jenna_gray)
len_of_pew=len(pew_gray)
```

```python
image_size_jk=jk_gray[1].shape
image_size_pew=pew_gray[1].shape
image_size_jenna=jenna_gray[1].shape
```

```python
image_size_pew
```

### Flattening the 2D array into 1D array

```python
flatten_size_jk=image_size_jk[0]*image_size_jk[1]
flatten_size_pew=image_size_pew[0]*image_size_pew[1]
flatten_size_jenna=image_size_jenna[0]*image_size_jenna[1]
```

```python
flatten_size_pew
```

### Converting the 2D into 3D array

```python
for i in range(len_of_jk):
    jk_gray[i]= np.ndarray.flatten(jk_gray[i].reshape(flatten_size_jk,1))
for i in range(len_of_pew):
    pew_gray[i]= np.ndarray.flatten(pew_gray[i].reshape(flatten_size_pew,1))
for i in range(len_of_jenna):
    jenna_gray[i]= np.ndarray.flatten(jenna_gray[i].reshape(flatten_size_jenna,1))
```

```python
jk_gray=np.dstack(jk_gray)
pew_gray=np.dstack(jenna_gray)
jenna_gray=np.dstack(pew_gray)
```

```python

pew_gray.shape
```

### Rollling the axis

```python
jk_gray=np.rollaxis(jk_gray,axis=2,start=0)
pew_gray=np.rollaxis(pew_gray,axis=2,start=0)
jenna_gray=np.rollaxis(jenna_gray,axis=1,start=0)
```

```python
pew_gray=np.rollaxis(pew_gray,axis=2,start=1)
jk_gray=np.rollaxis(jk_gray,axis=2,start=1)
```

```python
jenna_gray.shape
```

### Reshaping the array into 2D array

```python
jk_gray=jk_gray.reshape(len_of_jk,flatten_size_jk)
pew_gray=pew_gray.reshape(len_of_pew,flatten_size_pew)
jenna_gray=jenna_gray.reshape(len_of_jenna,flatten_size_jenna)
```

```python

jenna_gray.shape
```

### Storing the values of the images in a dataframe

```python
jk_data=pd.DataFrame(jk_gray)
jenna_data=pd.DataFrame(jenna_gray)
pew_data=pd.DataFrame(pew_gray)
```

```python
pew_data
```

### Adding label to the dataframe

```python
jk_data["label"]="Yogesh"
pew_data["label"]="Scarlet"
jenna_data["label"]="Brad Pitt"
```

```python
jenna_data
```

### Concatenating the 3 image sets into one dataframe

```python
actor_1=pd.concat([jk_data,pew_data])
```

```python
actor=pd.concat([actor_1,jenna_data])
```

```python
actor
```

#### Importing shuffle function and shuffle the dataframe

```python
from sklearn.utils import shuffle
```

```python
actor_shuffle = shuffle(actor).reset_index()
```

```python
actor_shuffle
```

### Dropping the Index

```python
actor_shuffle=actor_shuffle.drop(['index'],axis=1)
```

```python
actor_shuffle
```

```python
actor_shuffle.shape
```

### Creating test and train datasets and creating train and 

```python
x = actor_shuffle.values[:,:-1]
```

```python
y= actor_shuffle.values[:,-1]
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

### Importing SVM and training the model

```python
from sklearn import svm
```

```python
clf=svm.SVC()
clf.fit(x_train,y_train)
```

```python
y_pred=clf.predict(x_test)
```

```python
y_pred
```

```python
for i in (np.random.randint(0,6,4)):
    predicted_images = (np.reshape(x_test[i],(512,512)).astype(np.double))
    plt.title('predicited label:{0}'.format(y_pred[i]))
    plt.imshow(predicted_images,interpolation='nearest',cmap='gray')
    plt.show()
```

```python
from sklearn import metrics
```

```python
accuracy=metrics.accuracy_score(y_test,y_pred)
```
